```python
from ipythonblocks import colors, BlockGrid
from random import randint
```

```python
# Draw rainbows!!!!
rainbow_colors = ['Red', 'Purple', 'Blue', 'Green', 'Yellow', 'Orange']
n_colors = len(rainbow_colors)
grid = BlockGrid(41, 5, block_size=30, lines_on=False)
for i in range(100):
    for block in grid:
        color_name = rainbow_colors[(block.row + block.col + i) % n_colors]
        block.rgb = colors[color_name]
    grid.flash()
grid.show()
```

```python
def inside_grid(ball_position, grid):
    return ball_position[0] >= 0 \
        and ball_position[1] >= 0 \
        and ball_position[0] <= grid.shape[0] - 1 \
        and ball_position[1] <= grid.shape[1] - 1

def next_move():
    return (randint(-1, 1), randint(-1, 1))

def add_position(pos1, pos2):
    return tuple(map(sum, zip(pos1, pos2)))

ball_position = (randint(1, grid.shape[0]), randint(1, grid.shape[1]))
grid = BlockGrid(41, 10, block_size=30, lines_on=False)
for i in range(100):
    move = next_move()
    while move != (0,0) and not inside_grid(add_position(ball_position, move), grid):
        move = next_move()
    ball_position = add_position(ball_position, move)
    for block in grid:
        #print (block.row, block.col)
        if (block.col, block.row) == ball_position:
            block.rgb = colors['White']
        else:
            block.rgb = colors['Black']
    grid.flash()
grid.show()
```
